```python
black_img = np.zeros((512,512,3), dtype=np.int16)
plt.imshow(black_img)
```

```python
cv.rectangle(black_img, (384,0), (510,150), (0,255,0), thickness=10)
plt.imshow(black_img)
```

```python
cv.rectangle(black_img, pt1=(200,200), pt2=(300,300), color=(0,0,255), thickness=15)
plt.imshow(black_img)
```

```python
cv.circle(black_img, center=(100,100), radius=50, color=(255,0,0), thickness=8)
plt.imshow(black_img)
```

```python
cv.line(black_img, pt1=(0,0), pt2=(512,512), color=(255,255,255), thickness=5)
plt.imshow(black_img)
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(black_img, text='Hello', org=(10,500), fontFace=font, fontScale=4, color=(255,255,0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black_img)
```

```python
black_img3=np.zeros(shape=(512,512,3), dtype=np.int32)
plt.imshow(black_img3)

cv.rectangle(black_img3, pt1=(10,10), pt2=(100,100), color=(0,0,255), thickness=5)
plt.imshow(black_img3)

cv.circle(black_img3, center=(100,400), radius=50, color=(255,0,0), thickness=8)
plt.imshow(black_img3)

cv.line(black_img3, pt1=(0,256), pt2=(511,256), color=(255,255,255), thickness=10)
plt.imshow(black_img3)

cv.ellipse(black_img3, center=(100,256), axes=(100,50), angle=0, startAngle=0, endAngle=360, color=(255,0,0), thickness=5)
plt.imshow(black_img3)

font = cv2.FONT_HERSHEY_SIMPLEX
cv.putText(black_img3, text='iloveopencv2', org=(1,150), fontFace=font, fontScale=1, color=(255,255,0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(black_img3)
```
